```python
dt = pd.read_csv('gurgaon-flats-raw.csv' )
```

```python
dt = dt.drop(columns = 'Unnamed: 0')
```

```python
# deleted the empty rows 
dt = dt[~dt['propertyId'].isna()]
```

```python
# we reseted the index drop true replace the old index by new one 
dt = dt.reset_index(drop = True)
```

```python
# we dont need the links and propertyId columns as they are of no use
dt.drop(columns = ['links','propertyId'],inplace = True)
```

```python
# convert all to the lowercase so that captial and smallcase society names are not treated as two different 
dt['society'] = dt['society'].str.lower()
```

```python
dt.sample(2)
```

```python
dt['price'].value_counts() # we need to convert the price in same units (some are in crore some in lakhs)        
```

```python
dt['price'] = dt['price'].str.replace('Crore','')
```

```python
dt['price']
```

```python
def convert(val):
    if val.split(' ')[1] == 'Lac':
        x = float(val.split(' ')[0])
        x = x/100
        return round(x,2)
    else:
        return round(float(val),2)
                
```

```python
dt['price'] = dt['price'].apply(convert)
```

```python
dt.sample(2)
```

```python
# created a new column of area per square feet
```

```python
dt['pricePerSqft'] = round((dt['price']*10000000)/dt['area'],2)
```

```python
# removing the Bedrooms and bathrooms and converting it to integer 
```

```python
dt['bedRooms'] = dt['bedRooms'].str.replace('Bedrooms','')
dt['bedRooms'] = dt['bedRooms'].str.replace('Bedroom','')
dt['bedRooms'] = dt['bedRooms'].astype(int)
```

```python
dt['bathRooms'] = dt['bathRooms'].str.replace('Bathrooms','')
dt['bathRooms'] = dt['bathRooms'].str.replace('Bathroom','')
dt['bathRooms'] = dt['bathRooms'].astype(int)
```

```python
dt.sample(2)
```

```python
dt['additionalRooms'].value_counts()
```

```python
dt['additionalRooms'] = dt['additionalRooms'].str.lower()
```

```python
dt['additionalRooms'] = dt['additionalRooms'].fillna('not available')
```

```python
dt['floorNum'] = dt['floorNum'].str.split(' ').str.get(0).str.replace('Ground','0').str.replace('Basement','-1').str.strip().str.extract(r'(-?\d+)')
```
