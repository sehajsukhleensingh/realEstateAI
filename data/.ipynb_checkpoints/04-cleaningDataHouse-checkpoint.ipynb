```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max.rows' , None)
pd.set_option('display.max.columns' , None)
```

```python
dt = pd.read_csv('gurgaon-houses-raw.csv')
```

```python
dt = dt.drop(columns = 'Unnamed: 0')
```

```python
dt['price'].value_counts()
```

```python
dt['price'] = dt['price'].str.replace('₹','').str.strip()
```

```python
dt['price']
```

```python
def priceC(val):
    
    val = val.split(' ')
    
    if len(val) == 2:
        if val[1] == 'Lac':
            val = float(val[0])
            val = round((val/100),2)
            return val
        elif val[1] == 'Cr':
            val = float(val[0])
            return val
            
    elif len(val) == 3:
        return np.nan 
```

```python
dt['price'] = dt['price'].apply(priceC)
```

```python
dt.sample(3)
```

```python
dt['society'] = dt['society'].str.replace('in','').str.lower()
```

```python
dt['propertyType'] = dt['propertyType'].str.lower()
```

```python
dt['bedRooms'].value_counts()
```

```python
dt['bedRooms'] = dt['bedRooms'].str.split(' ').str.get(0).str.strip().astype(float)
dt['bathRooms'] = dt['bathRooms'].str.split(' ').str.get(0).str.strip().astype(float)
```

```python
dt.head(2)
```

```python
dt['address'] = dt['address'].str.replace('in','').str.lower()
```

```python
dt['additionalRooms'] = dt['additionalRooms'].str.lower()
dt['additionalRooms'] = dt['additionalRooms'].fillna('not available')
```

```python
dt.head(2)
```

```python
dt['floorNum'].value_counts()
```

```python
dt['floorNum'] = dt['floorNum'].str.split(' ').str.get(0).astype(float)
```

```python
dt['facing'] = dt['facing'].str.lower()
```

```python
dt['balcony'].value_counts()
```

```python
dt['balcony'] = dt['balcony'].str.strip().str.split(' ').str.get(0).str.replace('No','0')
```

```python
dt.to_csv('gurgaon-houses-cleaned-v1.csv',index = False)
```
