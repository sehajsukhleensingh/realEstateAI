```python
import pandas as pd
import numpy as np
```

```python
dt1 = pd.read_csv('links-gurgaon-homes.csv')
dt2 = pd.read_csv('data-homes.csv')
```

```python
dt1 = dt1.drop(columns = 'Unnamed: 0')
```

```python
dt1 = dt1.rename(columns = {'link':'links'})
dt1
```

```python
dt2 = dt2.drop(columns = ['Unnamed: 0'	, 'index'])
```

```python
dt2
```

```python
dt = dt1.merge(dt2,how='left',on = 'links')
```

```python
dt
```

```python
dt.duplicated().sum()
```

```python
dt = dt[~dt.duplicated()]
```

```python
dt.duplicated().sum()
```

```python
dt = dt[~dt['propertyId'].isna()]
```

```python
dt
```

```python
dt = dt.reset_index(drop  = True)
```

```python
dt
```

```python
dt.to_csv('gurgaon-houses-raw.csv')
```
