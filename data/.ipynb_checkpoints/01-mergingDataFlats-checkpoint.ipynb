```python
import pandas as pd
import numpy as np
```

```python
dt1 = pd.read_csv('99acres-gurgaon.csv')
dt2 = pd.read_csv('dt.csv')
```

```python
dt1.head() 
```

```python
dt1 = dt1.drop(columns = 'Unnamed: 0')
```

```python
dt1 = dt1[~dt1['links'].duplicated(keep = 'first')] # removing the duplicate data 
```

```python
dt1.duplicated().sum()
```

```python
dt2.head()
```

```python
dt2 = dt2.drop(columns = 'Unnamed: 0')
```

```python
dt2.head()
```

```python
dt2.duplicated().sum()
```

```python
dt2 = dt2[~dt2.duplicated(keep  = 'first')]
```

```python
dt2.duplicated().sum()
```

```python
dt = dt1.merge(dt2 , on = 'links' , how = 'left') #merging the dataframes on the link as index 
```

```python
dt
```

```python
dt.head(2)
```

```python
dt['society'] = dt['society'].str.replace('in','')
```

```python
dt['address'] = dt['address'].str.replace('\n',' ')
```

```python
dt.to_csv('gurgaon-flats-raw.csv')
```
