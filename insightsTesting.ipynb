    we have coef from a model which is trained on standardised X and Y log , to convert them back to original scale we use the formula

    The absolute change in price $\Delta Y$ for an increase of $\Delta X$ in a feature is given by:
    
$\Delta Y = (1 + Y_{\text{baseline}}) \cdot \left[ \exp\left(\beta_s \cdot \frac{\sigma_{Y_{\text{log}}}}{\sigma_X} \cdot \Delta X \right) - 1 \right]$
    
    Where:  
- $\beta_s$ = standardized coefficient from the linear model  
- $\sigma_{Y_{\text{log}}}$ = standard deviation of log-transformed target $Y$ (i.e., $\log(1+Y)$)  
- $\sigma_X$ = standard deviation of the feature $X$  
- $\Delta X$ = the increase in the feature for which we want to measure impact  
- $Y_{\text{baseline}}$ = baseline value of the target (can use mean or median price in original units)

```python
Ymean = dt['Y'].mean()
stdYlog = Y.std()

stdCols = {}

for i in dt.columns:
    stdCols[i] = dt[i].std()
```

```python
stdCols
```

```python
unstdCoef = {}
for i in stdCols:
    coefVal = coef[coef['features'] == i ]['coef'].values
    if coefVal.size == 0:
        pass
    else:
        var =  np.expm1(coefVal * (stdYlog / stdCols[i]))*(1+Ymean)
        unstdCoef[i] = var
```

```python
unstdCoef
```

```python
coefUnstd = pd.DataFrame(unstdCoef)
```

```python
coefUnstd = coefUnstd.stack().reset_index()
coefUnstd.drop(columns = 'level_0' , inplace = True)
coefUnstd.rename(columns = {'level_1':'features',0:'coef'} , inplace = True)
```
